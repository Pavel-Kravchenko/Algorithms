```python
class ListBlock:
    def __init__(self, value = None, next = None):
        self.value = value
        self.next = next

class LinkedList:
    def __init__(self):
        self.first = None
        self.last = None
        self.length = 0
    
    def add(self, elem):
        self.length += 1
        if self.first == None:
            self.last = self.first = ListBlock(elem)
        else:
            self.last.next = self.last = ListBlock(elem)
            
    def Del_end(self):
        curr = self.first
        while curr.next != None:
            if curr.next == self.last: 
                break
            curr = curr.next
        curr.next = None
        self.last = curr
    
    def search_elm_by_value(self, elm):
        if (self.first == None):
            return -1
        curr = self.first
        calc = 0
        while curr != None:
            if curr.value == elm:
                return calc
            curr = curr.next
            calc += 1
        if curr == None:
            return -1
        
    def search_elm_by_index(self, index):
        if (self.first == None):
            return -1
        curr = self.first
        calc = 0
        while curr != None:
            if calc == index:
                return curr.value
            curr = curr.next
            calc += 1
        if curr == None:
            return -1
        
    def Del_elm_by_index(self,index):
        if (self.first == None):
            return -1
        curr = self.first
        count = 0
        if index == 0:
            self.first = self.first.next
            return 
        while curr != None:
            if count == index:
                if curr.next == None:
                    self.last = curr
                old.next = curr.next 
                break
            old = curr  
            curr = curr.next
            count += 1
            
    def Del_elm_by_value(self,data):
        if (self.first == None):
            return -1
        else:
            curr  = self.first
            prev = None
            while curr.value != data and curr.next != None:
                prev = curr
                curr = curr.next

            if curr.value == data:
                if curr == self.first:
                    if curr.next == None:
                        self.first = None
                    else:
                        self.first = curr.next
                else:
                    if curr.next == None:
                        prev.next = None
                    else:
                        prev.next = curr.next
                return
            else:
                return -1
        
    def __str__(self):
        if self.first != None:
            curr = self.first
            out = 'LinkedList [ ' +str(curr.value) +' '
            while curr.next != None:
                curr = curr.next
                out += str(curr.value) + ' '
            return out + ']'
        else:
            return 'LinkedList []'
        
    def reverse(self):
        curr = self.first 
        prev = None       
        while curr != None:
            nextt =  curr.next  
            curr.next = prev
            prev = curr
            curr = nextt
        self.first = prev 
```

2. Написать QSort с процедурой Partition работающей с использованием O(1) дополнительной памяти. Это разбиралось на семинаре. Использовать срезы запрещено. Использовать операции pop, append - запрещено. (2.5 балла). Использовать рекурсию можно. 
     

```python
import random # Просто собираем всё в списки и сортируем рекурсивно

def quicksort(alist):
    if len(alist) <= 1:
        return alist
    else:
        q = random.choice(alist)
        s_nums = []
        m_nums = []
        e_nums = []
        for n in alist:
            if n < q:
                s_nums.append(n)
            elif n > q:
                m_nums.append(n)
            else:
                e_nums.append(n)
    return quicksort(s_nums) + e_nums + quicksort(m_nums)
```

```python
import random

def quicksort(alist, fst, lst):
    if fst >= lst: 
        return -1
    # shuffle(alist)
    i, j = fst, lst
    pivot  = random.choice(alist)
    #print(pivot)
    # Partition
    while i < j:
        while alist[i] < pivot: 
            #print(i, alist[i])
            i += 1
        while alist[j] > pivot: 
            #print(j, alist[j])
            j -= 1
        if i <= j:
            alist[i], alist[j] = alist[j], alist[i]
            #print(alist[i], alist[j], "=", alist[j], alist[i])
            i, j = i + 1, j - 1
    #print(alist)
    quicksort(alist, fst, j)
    quicksort(alist, i, lst)
    return alist
```

```python
import random # Просто собираем всё в списки и сортируем рекурсивно с момощью list comprechension 

def quicksort(nums):
    if len(nums) <= 1:
        return nums
    else:
        q = random.choice(nums)
    l_nums = [n for n in nums if n < q]
 
    e_nums = [q] * nums.count(q)
    b_nums = [n for n in nums if n > q]
    return quicksort(l_nums) + e_nums + quicksort(b_nums)
```

```python
alist = [1,23,34,45,4,5,6,7,8,56,78,90,9,88,100,2,3,10,11,12,13,14,15,16,17,18,111,11111]
print(quicksort(alist))
```
